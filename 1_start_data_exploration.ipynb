## Final Data 만들기

[변수 구성]  

date, bus_route_id, station_code, station_name, latitude, logitude,  
6-7 ride ... 11-12 ride, ride_total  
6-7 takeoff ... 11-12 takeoff, takeoff_total  
type_general_ride, type_others_ride, type_others_ride, type_others_ride, holiday, weather, bus_type

```python
final_data = pd.DataFrame()
final_data['date'] = data['date']
final_data['bus_route_id'] = data['bus_route_id']
final_data['station_code'] = data['station_code']
final_data['station_name'] = data['station_name']
final_data['latitude'] = data['latitude']
final_data['longitude'] = data['longitude']
final_data['6~7_ride'] = data['6~7_ride']
final_data['7~8_ride'] = data['7~8_ride']
final_data['8~9_ride'] = data['8~9_ride']
final_data['9~10_ride'] = data['9~10_ride']
final_data['10~11_ride'] = data['10~11_ride']
final_data['11~12_ride'] = data['11~12_ride']
final_data['6~7_takeoff'] = data['6~7_takeoff']
final_data['7~8_takeoff'] = data['7~8_takeoff']
final_data['8~9_takeoff'] = data['8~9_takeoff']
final_data['9~10_takeoff'] = data['9~10_takeoff']
final_data['10~11_takeoff'] = data['10~11_takeoff']
final_data['11~12_takeoff'] = data['11~12_takeoff']
final_data['ride_total'] = data['ride_total']
final_data['takeoff_total'] = data['takeoff_total']
final_data.head()
```

```python
final_data.tail()
```

### 공휴일 변수 만들기

```python
holiday_date = []
holiday_date = ['2019-09-01', '2019-09-07','2019-09-08','2019-09-12','2019-09-13','2019-09-14','2019-09-15','2019-09-21','2019-09-22','2019-09-28','2019-09-29']

final_data['holiday'] = np.nan
```

```python
holiday = []

for i in range(len(final_data)):
    if final_data['date'][i] in holiday_date:
        holiday.append(0)
    else:
        holiday.append(1)
```

```python
final_data['holiday'] = holiday
final_data.head()
```

```python
final_data.to_csv('final_data.csv')
```

## 승객 유형별 승하차인원 변수 추가한 데이터 확인하기
**(made from  R)**

일반 승객과 청소년 승객의 승하차인원 = type_general_geton / type_general_getoff  
그 외의 나머지 승객의 승하차인원 = type_others_geton / type_others_getoff

```python
final_category = pd.read_csv('final_category.csv', encoding='ANSI')
final_category.head()
```

```python
final_category['type_general_geton'] = final_category['type_general_geton'].fillna(0)
final_category['type_other_geton'] = final_category['type_other_geton'].fillna(0)
final_category.head()
```

```python
final_category[final_category['type_other_geton'] + final_category['type_general_geton'] != final_category['ride_total']]
```

```python
final_category[final_category['type_other_getoff'] + final_category['type_general_getoff'] != final_category['takeoff_total']]
```

하차태그를 하지 않은 승객이 있어서  
일반 승객 하차인원 + 나머지 승객 하차인원 값이 총 하차인원과 다를 수 있다.
